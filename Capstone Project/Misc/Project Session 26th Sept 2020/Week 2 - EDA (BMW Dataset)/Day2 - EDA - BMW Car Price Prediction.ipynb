### Price pattern w.r.t. time analysis

```python
## converting time columns as time type
df.registration_date = pd.to_datetime(df.registration_date)
df.sold_at = pd.to_datetime(df.sold_at)
```

```python
###  Exacting time components from time columns
df['reg_year'] = df.registration_date.dt.year
df['reg_month'] = df.registration_date.dt.month
df['reg_weekday'] = df.registration_date.dt.weekday
df['reg_weekday_yr'] = df.registration_date.dt.weekofyear

df['sold_year'] = df.sold_at.dt.year
df['sold_month'] = df.sold_at.dt.month
df['sold_weekday'] = df.sold_at.dt.weekday
df['sold_weekday_yr'] = df.sold_at.dt.weekofyear

df['diff_reg_sold'] = df.sold_at - df.registration_date
df.diff_reg_sold = df.diff_reg_sold.apply(lambda x: str(x).split()[0])

# year difference between sold year and registration year
df['year_diff']  = df.sold_year - df.reg_year 
```

### model_key impact on price and relation with engine_power

```python
### relation of model_key with engine_power and price (if any)
engine_power = []
for m_key in df.model_key:
    x = df[df['model_key']==m_key]
    engine_power.append(sum(x['engine_power']/len(x)))

plt.figure(figsize=(10,20))
df['engine_power_per_unit'] = engine_power
df.sort_values('engine_power_per_unit', inplace = True, ascending = False)
sns.barplot(x=df.engine_power,y=df.model_key)
plt.xlabel("Engine Power")
plt.ylabel("Model Key")
plt.title("Engine Power a/c to Model")
plt.show()
```

```python
## binning can be peformed to indicate relation of engine_power with model_key
```

```python
### relation of model_key with price (if any)
price_PU = []
for m_key in df.model_key:
    x = df[df['model_key']==m_key]
    price_PU.append(sum(x['price']/len(x)))

plt.figure(figsize=(10,20))
df['price_PU'] = price_PU
df.sort_values('price_PU', inplace = True, ascending = False)
sns.barplot(x=df.price_PU,y=df.model_key)
plt.xlabel("Price per Unit")
plt.ylabel("Model Key")
plt.title("Price a/c to Model")
plt.show()
```

```python
## binning can be peformed to indicate relation of price per unit with model_key
```

### impact of fuel type

```python
fig, ax = plt.subplots(1,2, figsize = (15,7))
sns.countplot(df.fuel, ax = ax[0])
sns.histplot(x = 'price', hue = 'fuel', data = df, ax = ax[1])
```

```python
from statistics import median
price_ft = []
fuel_type = []
for each_fuel in df.fuel.unique():
    x = df[df.fuel==each_fuel]
    price_ft.append(median(x['price']))
    fuel_type.append(each_fuel)
 
fig, ax = plt.subplots(1,3, figsize = (20,7))
sns.countplot(df.fuel, ax = ax[0])
sns.histplot(x = 'price', hue = 'fuel', data = df, ax = ax[1]) 
sns.barplot(x = price_ft, y =fuel_type,   ax = ax[2])
ax[2].set_xlabel('Price_per_unit')
ax[2].set_xlabel('Fuel Types')
```

### impact of car type

```python
### relation of car_type with price (if any)
price_PU = []
for c_key in df.car_type:
    x = df[df['car_type']==c_key]
    price_PU.append(sum(x['price']/len(x)))

fig, ax = plt.subplots(1,2,figsize=(15,6))
df['AvgPrice_CT'] = price_PU
df.sort_values('AvgPrice_CT', inplace = True, ascending = False)
sns.barplot(x=df.AvgPrice_CT,y=df.car_type, ax = ax[1])
ax[1].set_xlabel("Avg. Price")
ax[1].set_ylabel("Car Type")
ax[1].set_title("Price a/c to Car type")

sns.countplot(df.car_type, ax = ax[0])
plt.show()
```
