```python
## dropping records with single missing values
df.card3.dropna(inplace = True)
df.card4.dropna(inplace = True)
df.card6.dropna(inplace = True)
```

```python
## type casting 
for each in df[['card3']].columns:
    df[each] = df[each].astype('int')
```

```python
## string comparison
df['complt_card'] =  df.card1.astype('str').astype('object') + df.card2.astype('str').astype('object') + df.card3.astype('str').astype('object') + df.card4.astype('str').astype('object') +df.card5.astype('str').astype('object')+ df.card6.astype('str').astype('object')
```

```python
df.head()
```

```python
x = df.groupby('complt_card')['isFraud'].count().reset_index()
y = df.groupby('complt_card')['isFraud'].nunique().reset_index()
z = df.groupby('complt_card')['isFraud'].unique().reset_index()
x = x.merge(y, on = ['complt_card']).merge(z, on = ['complt_card'] )
print(x[x.isFraud_y > 1].shape)  ## a transaction resulting to both the cases: fraud and none-fraud
print(x[x.isFraud_y == 1].shape)
x[x.isFraud_y > 1]
```
